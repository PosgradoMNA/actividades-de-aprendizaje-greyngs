1. **Fundamentos de bases de datos y para ciencia de datos.**

Los datos son el alma de la ciencia de datos, todo gira alrededor de los datos, estos datos pueden estar almacenados en diferentes partes, de esta forma las bases de datos son un conjunto de datos almacenados de forma estructurada y lógicamente relacionadas, a estas se les conoce como bases de datos relacionales aunque también existen bases de datos no relacionales.
Para manipular las bases de datos relacionales se utiliza SQL (Structured Query Language) con el cual se puede agregar, borrar, actualizar o cambiar la estructura de la db, este lenguaje es muy popular por su flexibilidad, facilidad de uso, la no redundancia y su fiabilidad.
Por otra parte las db NoSQL se comportan de cierta forma como diccionarios de python, donde se accede a los datos por medio de llaves.

2. **Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.**

Un data warehouse se entiende como un sistema para la gestión de datos diseñado para darle más soporte a las actividades de BI y al almacenamiento de grandes volúmenes de datos. Se centralizan las fuentes de datos para tener en un solo lugar el manejo de los datos. Estos almacenes de datos pueden estar en la nube y los sistemas modernos buscan hacer más eficiente el flujo de los datos, algo difícil debido al volumen de estos. En conclusión es un lugar donde se almacenan grandes cantidades de datos que busca tener centralizadas las fuentes y facilitar el trabajo para la toma de decisiones de negocio.
